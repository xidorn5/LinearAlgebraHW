```python
from sympy import *
import numpy as np
init_printing(use_unicode=True)


'''
Program to anayse the voting data of UN member countries.
Input: Voting record in a text file
Output: The top ten most opposed pairs of countries (in descending order of opposition) 
        and the topmost pair of countries with the greatest agreements
        
Date: 07/14/2016

'''

def min_sort(arr, n):
    '''
    Returns the elements of an array in ascending number
    Takes two arguments:
        arr - the array
        n - the number of elements of the array to be returned
    '''
    indices = arr.ravel().argsort()[:n]
    indices = (np.unravel_index(i, arr.shape) for i in indices)
    return [i for i in indices]
    
def max_sort(arr, n):
    '''
    Returns the elements of an array in ascending number
    Takes two arguments:
        arr - the array
        n - the starting point for the elements to be returned
    '''
    indices = arr.ravel().argsort()[-n:]
    indices = (np.unravel_index(i, arr.shape) for i in indices)
    return [i for i in indices]

FILE_NAME = ("/Users/divinity/Linear Algebra/UN_voting_data.txt")

with open(FILE_NAME) as textFile:
    lines = textFile.readlines()

text = open(FILE_NAME)

new_text = text.read().split()  
country = []
votes = []

#collate the country labels into a list
for word in new_text:
    if len(word) > 1 and word[0] != "-":
        country.append(word)    

#collate the voles into another list and convert the numbers from strings to integers
for i in range(len(lines)):
    votes.append(lines[i].split())
    votes[i].remove(country[i])
    for j in range (len(votes[i])):
        votes[i][j] = int(votes[i][j])
        
#Matrix operations. To position the countries as the columns of A, transpose B.
B = np.matrix(votes)
A = B.T
M = A.T * A

# Which pair of countries are most opposed?
arrM = np.array(M)

min_indices = np.where(arrM == arrM.min())
print ("\n\nA. The most opposed pairs of countries are:")
print ("{} and {}".format(country[min_indices[0][0]], country[min_indices[0][1]]))



#--------ANSWERS TO SPECIFIC QUESTIONS-----------
#------------------------------------------------


#What are the ten most opposed pairs of countries?

#The countries in the results are duplicated since it is a symmetric matrix. 
min_indices = min_sort(arrM,20)
min_countries =[]


# Duplicates: so step through iteration in 2's instead of 1's
for i in range(0, len(min_indices), 2):
        min_countries.append(( country[min_indices[i][0]],country[min_indices[i][1]]))

print("\n\nB. The ten most opposed pairs of countries (in descending order) are:")
for i in range(len(min_countries)):
    print ("{}.  {} and {}".format( i+1, min_countries[i][0], min_countries[i][1]))
 
 
# Which pair of distinct countries are in the greatest agreement (have the most positive dot-product)?
max_indices = max_sort(arrM, -1)
max_indices.reverse()

for item in max_indices:
    if item[0] != item[1]:
        print("\n\nC. The distinct pair of countries with the greatest agreement are:")
        print ("{} and {}".format(country[item[0]],country[item[1]]))
        break
```
